#Parte 2: Medindo Latência de Memória


```python
%%writefile sequencial.c
/*
Este é um código em C projetado para medir o tempo necessário para
inicializar sequencialmente um array de inteiros. Ele aloca
memória para o array, preenche-o com valores e calcula o tempo decorrido
*/

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define SIZE 100000000

int main() {
    int *array = malloc(sizeof(int) * SIZE);

    clock_t start = clock();
    for (long i = 0; i < SIZE; i++) {
        array[i] = i;
    }
    clock_t end = clock();

    double time_spent = (double)(end - start) / CLOCKS_PER_SEC;
    printf("Tempo sequencial: %f segundos\n", time_spent);

    free(array);
    return 0;
}
```

```python
! if [ ! -f sequencial ]; then gcc sequencial.c -o sequencial; fi;
!./sequencial
```

```python
%%writefile aleatorio.c
/*
Este é um código em C projetado para medir o tempo necessário para
inicializar sequencialmente um array grande de inteiros. Ele aloca
memória para o array, preenche-o com valores e calcula o tempo decorrido
*/

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define SIZE 100000000

int main() {
    int *array = malloc(sizeof(int) * SIZE);

    clock_t start = clock();
    for (long i = 0; i < SIZE; i++) {
        array[rand() % SIZE] = i;
    }
    clock_t end = clock();

    double time_spent = (double)(end - start) / CLOCKS_PER_SEC;
    printf("Tempo sequencial: %f segundos\n", time_spent);

    free(array);
    return 0;
}
```

```python
! if [ ! -f aleatorio ]; then gcc aleatorio.c -o aleatorio; fi;
!./aleatorio
```
